```python
import pandas as pd
import matplotlib.pyplot as py
import numpy as np
```

```python
df=pd.read_csv("spotify_data.csv")
df.head()
```

```python
missing_values=df.isnull()
missing_count=missing_values.sum()
```

```python
print(missing_count)
```

```python
df['Daily'] = df['Daily'].fillna(df['Daily'].mean(skipna=True))
```

```python
import seaborn as sns
sns.scatterplot(x='Streams', y='Daily', data=df)
py.title('Streams vs. Daily')
py.xlabel('Streams')
py.ylabel('Daily')
py.show()
```

```python
sns.lineplot(x='Streams', y='Daily', data=df)
py.title('Daily Streams vs. Total Streams')
py.xlabel('Streams')
py.ylabel('Daily')
py.show()
```

```python
sns.histplot(x='Streams', y='Daily', data=df, bins=20)
py.title('Distribution of Daily Streams by Total Streams')
py.xlabel('Streams')
py.ylabel('Daily')
py.show()
```

```python
sns.jointplot(x='Streams', y='Daily', data=df)
```

```python
df['popularity_score'] = (df['Daily'] * 0.5) + (df['Streams'] * 0.5)
most_popular_song = df.sort_values(by='popularity_score', ascending=False).iloc[0]

print("Most popular song:")
print(most_popular_song)
```

```python
df['popularity_score'] = (df['Daily'] * 0.5) + (df['Streams'] * 0.5)
least_popular_song = df.sort_values(by='popularity_score', ascending=True).iloc[0]

print("Least popular song:")
print(least_popular_song)
```

```python
song_with_max_daily = df.sort_values(by='Daily', ascending=False).iloc[0]
song_with_min_daily = df.sort_values(by='Daily', ascending=True).iloc[0]

print("Song with maximum daily streams:")
print(song_with_max_daily)

print("Song with minimum daily streams:")
print(song_with_min_daily)
```

```python
song_with_max_streams = df.sort_values(by='Streams', ascending=False).iloc[0]
song_with_min_streams = df.sort_values(by='Streams', ascending=True).iloc[0]

print("Song with maximum total streams:")
print(song_with_max_streams)

print("Song with minimum total streams:")
print(song_with_min_streams)
```

```python
songs = [row['Songs & Artist'].split("-")[1] for index, row in df.iterrows()]
artists = [row['Songs & Artist'].split("-")[0] for index, row in df.iterrows()]

songs_df = pd.DataFrame({'Songs': songs})
artists_df = pd.DataFrame({'Artists': artists})
```

```python
print(songs_df.head())
print(artists_df.head())
```

```python
songs_markdown = songs_df.to_markdown(index=False)
artists_markdown = artists_df.to_markdown(index=False)

print("Songs:")
print(songs_markdown)

print("\nArtists:")
print(artists_markdown)
```

```python
daily_df = df['Daily']
streams_df =df['Streams']

print(daily_df)
print(streams_df)
```

```python

def process_songs_markdown(markdown_text):
    lines = markdown_text.splitlines()
    songs_list = [line.strip() for line in lines if line.strip()]
    return songs_list

songs_list = process_songs_markdown(songs_markdown)


songs_df = pd.DataFrame({'Songs': songs_list})
```

```python
combined_df = pd.concat([songs_df, artists_df, daily_df, streams_df], axis=1)
combined_df.columns = ['Songs', 'Artists', 'Daily', 'Streams']

print(combined_df)
```
